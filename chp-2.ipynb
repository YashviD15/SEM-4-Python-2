```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
x=pd.read_csv("auto-mpg.csv")
y=pd.DataFrame(x)
y
```

```python

plt.stackplot(x["mpg"],x["acceleration"],x["origin"],labels=["Y","X"])
plt.legend()
plt.show()
```

```python
plt.fill_between(y["mpg"],y["acceleration"],where=y["cylinders"]>4,color="red",interpolated=True)
plt.fill_between(y["mpg"],y["acceleration"],where=y["cylinders"]<4,color="blue",interpolated=True)
plt.show()
```
